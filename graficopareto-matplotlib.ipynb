```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plt.rcParams.update({'font.size':14})
```

```python
df = pd.DataFrame({'tipo defeito':['nao selagem topo','nao selagem fundo',
                                  'nao selagem lateral', 'impressao borrada',
                                  'caixa amassada', 'outros'],
                  'frequencia':[55,38,22,8,7,6]})
```

```python
df
```

```python
df['frequencia'].cumsum()
```

```python
df['frequencia'].sum()
```

```python
df['cum (%)'] = df['frequencia'].cumsum()/df['frequencia'].sum()*100
```

```python
df
```

```python
fig,ax1 = plt.subplots(figsize = (12,6))

ax1.set_title('Pareto')

color1 = 'red'
color2 = 'black'
ax1.set_ylabel('Frequncia',color = color2)

ax1.bar(df['tipo defeito'], df['frequencia'],color = color1)
ax1.tick_params(axis = 'y',labelcolor = color2)

ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('%',color = color2) 

ax2.plot(df['tipo defeito'], df['cum (%)'],color = color2,marker = 's',markersize = 8, linestyle = '-')

ax2.tick_params(axis = 'y',labelcolor = color2)
ax2.set_ylim([0,120])

for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
    
plt.savefig('GraficoPareto.png',format='png',dpi = 600, bbox_inches = 'tight')
plt.show()
```
